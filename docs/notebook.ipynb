# SIR Epidemic Model Explained

## Introduction

The SIR model is a classical compartmental model in epidemiology. It divides the population into three groups:

- **S (Susceptible)**: Individuals who can catch the disease
- **I (Infected)**: Individuals who are sick and can transmit the disease
- **R (Recovered)**: Individuals who have recovered and are immune

## Differential Equations

The model is described by three ordinary differential equations:

$$
\frac{dS}{dt} = -\beta S I
$$

$$
\frac{dI}{dt} = \beta S I - \gamma I
$$

$$
\frac{dR}{dt} = \gamma I
$$

Where:
- $\beta$ = transmission rate (contacts × probability of infection)
- $\gamma$ = recovery rate (1 / infectious period)

## Basic Reproduction Number

$$R_0 = \frac{\beta}{\gamma}$$

- $R_0 > 1$: Epidemic occurs
- $R_0 = 1$: Disease becomes endemic
- $R_0 < 1$: Disease dies out

```python
# Import libraries
import numpy as np
import matplotlib.pyplot as plt
from sir_model import run_sir_simulation, calculate_R0, calculate_peak_infected

# Run simulation
beta = 0.5
gamma = 0.2
S0 = 990
I0 = 10
R0 = 0
t_max = 100
steps = 500

df = run_sir_simulation(beta, gamma, S0, I0, R0, t_max, steps)

# Plot results
fig, ax = plt.subplots(figsize=(12, 6))
ax.plot(df['time'], df['Susceptible'], label='Susceptible', lw=2)
ax.plot(df['time'], df['Infected'], label='Infected', lw=2)
ax.plot(df['time'], df['Recovered'], label='Recovered', lw=2)
ax.set_xlabel('Time (days)')
ax.set_ylabel('Population')
ax.set_title(f'SIR Model Simulation (β={beta}, γ={gamma}, R₀={calculate_R0(beta, gamma):.2f})')
ax.legend()
ax.grid(True, alpha=0.3)
plt.show()

# Display statistics
peak_infected, peak_time = calculate_peak_infected(df)
print(f"Peak infected: {peak_infected:.0f} at day {peak_time:.1f}")
print(f"Final recovered: {df['Recovered'].iloc[-1]:.0f}")
print(f"R₀ = {calculate_R0(beta, gamma):.3f}")
```

## Experiment with Different Parameters

Try changing β and γ to see how they affect the curve shape.

```python
# Compare multiple scenarios
scenarios = [
    (0.5, 0.2, "R₀ = 2.5 (Epidemic)"),
    (0.3, 0.3, "R₀ = 1.0 (Borderline)"),
    (0.2, 0.4, "R₀ = 0.5 (Dying out)")
]

fig, axes = plt.subplots(1, 3, figsize=(15, 4))

for idx, (beta, gamma, title) in enumerate(scenarios):
    df = run_sir_simulation(beta, gamma, 990, 10, 0, 100, 500)
    axes[idx].plot(df['time'], df['Susceptible'], label='S')
    axes[idx].plot(df['time'], df['Infected'], label='I')
    axes[idx].plot(df['time'], df['Recovered'], label='R')
    axes[idx].set_title(title)
    axes[idx].set_xlabel('Days')
    axes[idx].legend()
    axes[idx].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```
